```python
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
import math
```

```python
# inputs from the user
sigma_xx=float(input('σ_x:'))
sigma_yy=float(input('σ_y:'))
tau_xy=float(input('τ_xy:'))
theta=float(input('θ:'))

# Radius of circle
R=math.sqrt(((sigma_xx-sigma_yy)/2)**2 + (tau_xy)**2)

# Center of circle
center=(sigma_xx+sigma_yy)/2

# evaluating the principal stress components
sigma_1=round(center+R,2)
sigma_2=round(center-R,2)


# new coordinates in rotated axis
tau_xy_new=round(((tau_xy*cos(4*theta*math.pi/360)+(sigma_yy-sigma_xx)*sin(4*theta*math.pi/360)/2)),2)
sigma_xx_new=round(((sigma_xx+sigma_yy)/2+(sigma_xx-sigma_yy)*cos(4*theta*math.pi/360)/2+tau_xy*sin(4*theta*math.pi/360)),2)
sigma_yy_new=round(((sigma_xx+sigma_yy)/2-(sigma_xx-sigma_yy)*cos(4*theta*math.pi/360)/2-tau_xy*sin(4*theta*math.pi/360)),2)

rad=np.linspace(0,360,360)*2*math.pi/360

# points for plotting the circle
sigma=center+R*np.cos(rad)
tau=R*np.sin(rad)

# things used in plotting
x=[sigma_xx,sigma_yy]
y=[tau_xy,-tau_xy]
a_new=[sigma_xx_new,sigma_yy_new]
b_new=[tau_xy_new,-tau_xy_new]
a=[center,sigma_xx,sigma_yy,sigma_1,sigma_2]
b=[0,tau_xy,-tau_xy,0,0]

# the plot
plt.figure(figsize=(8,8))
plt.plot(a,b,'ro')
plt.plot(x,y,sigma,tau)
plt.plot(a_new,b_new)
axis('equal')

# showing the x and y axis
plt.axhline(y=0,color='lightgrey')
plt.axvline(x=0,color='lightgrey')
plt.xlabel('σ')
plt.ylabel('τ')

# showing the coordinates
for x,y in zip(a,b):
    plt.text(x,y,'({},{})'.format(x,y))  
for x,y in zip(a_new,b_new):
    plt.text(x,y,'({},{})'.format(x,y))
    
# print statements
print('..................')
print('Radius of Mohr Circle =',round(R,2))
print('..................')
print('Principal stress values are : σ_1 =',sigma_1,'and σ_2 =',sigma_2)
print('..................')
print('In rotated system:')
print('σ_x =',sigma_xx_new,', σ_y =',sigma_yy_new,', τ_xy =',tau_xy_new)
```
